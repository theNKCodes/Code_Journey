```python
data = [[-1,2],[-0.5,6],[0,10],[1,18]]
```

```python
scaler = MinMaxScaler()
```

```python
print(scaler.data_min_)
print(scaler.data_max_)
```

```python
scaler.fit(data)
```

```python
sd=scaler.transform(data)
sd
```

```python
cp = scaler.fit_transform(data)
cp
```
